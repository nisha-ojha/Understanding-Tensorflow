```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, Flatten, Input
```

```python
from tensorflow.keras.utils import to_categorical #this is used for one hot encoding
```

```python
from tensorflow.keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
x_train = x_train/255.0  #Each pixel is an integer between 0 and 255. 0 = black, 255 = white, Scales all pixel values to the range 0 to 1

# 0 / 255 = 0.0

#255 / 255 = 1.0
```

```python
x_train = x_train/255.0
```

```python
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
model = Sequential([
    Flatten(input_shape=(28, 28)),   
    Dense(128, activation="relu"),
    Dense(10, activation="softmax")
])
```

```python
#compile the model where the model gets trained
#Use Adam to update weights, measure error using categorical crossentropy, and track accuracy


model.compile(
    optimizer="adam", #Adaptive Moment Estimation----> how the model updates weights
    loss="categorical_crossentropy", #Measures how wrong the model’s predictions are
#Output has multiple classes
#Labels are one-hot encoded


    metrics=["accuracy"] #Tells TensorFlow what to report during training
)
```

```python
model.fit(x_train, y_train, epochs=5, batch_size=32)
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test accuracy: {test_acc:.4f}")
```
